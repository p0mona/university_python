## Task 1

Read the topic "Basics of reading and writing" from lecture 10.

1. Create a text file .txt on the disk containing 10 lines of some text (a few words per line).

2. Write a function that creates a list of texts and returns it. Each element of this list should contain **one line** from the file whose path is a parameter of the function.
- Be sure to use the exception mechanism when opening the file that handles error situations (`except Exception as ...`).
- Use the `read` method to read the entire file and the `split` method to split the text into lines.
- Make sure to close the file.

3. Test the function on the created file. Check whether the result of the function is a list of texts.

````python
# Solution of exercise 1
def text_output(s):
    try:
        with open(s, 'r') as file:
            lines = file.read().splitlines()
            return lines
    except Exception as e:
        print(f"Error reading file: {e}")
        return []

s = 'lab.txt'
try:
    lines_list = text_output(s)
    if lines_list:
        for line in lines_list:
            print(line)
except Exception as e:
    print(f"An error occurred: {e}")
````

Creating a list of lines is inefficient when the files are large. In such cases, it's better to use other methods for reading.

Let's assume we want to print all the words containing a given letter for each line. For example, for a file with the following content:
```
Jola
ma
kota, jeża i jaskółkę.
Zażółć gęślą jaźń.
```
and the letter `'j'`, the result should be:
```
1: Jola

3: jeża jaskółkę
4: jaźń
```

Write a function that does this and test it. The function's parameters should be the file path and the character.

Hint: (read the entire first answer on https://stackoverflow.com/questions/3277503/how-to-read-a-file-line-by-line-into-a-list and choose the correct solution that does not create a list).

To remove punctuation marks, use the following function (source: https://blog.enterprisedna.co/python-remove-punctuation-from-string/):
```python
import string

def remove_punctuation(input_string):
    """https://blog.enterprisedna.co/python-remove-punctuation-from-string/"""
    translator = str.maketrans('', '', string.punctuation + "\n")
    return input_string.translate(translator)
```

````python
# Solution of exercise 2
import string

def remove_punctuation(input_string):
    translator = str.maketrans('', '', string.punctuation + "\n")
    return input_string.translate(translator)

def find_s(f, l):
    try:
        with open(f, 'r', encoding='utf-8') as file:
            line_number = 0
            for line in file:
                line_number += 1
                line = remove_punctuation(line)
                words = line.split()
                words_with_letter = [word for word in words if l.lower() in word.lower()]
                result_line = f"{line_number}: {' '.join(words_with_letter)}"
                print(result_line)
    except FileNotFoundError:
        print(f"File '{f}' not found.")
f = "lab.txt"
l = input()
find_s(f, l)


````

## Task 3

1. Read the topic on CSV files in lecture 10 and the information from https://earthly.dev/blog/csv-python/.

1. Write a function that prints the first `n` lines of a CSV file using the `csv.reader` method. The parameters of the function should be the path to the file and `n`.

1. Download the file https://raw.githubusercontent.com/adamgordonbell/python-examples/main/readcsv/bwq.csv from Kaggle (while you're at it, learn what Kaggle is and what the prizes are in the competitions).

1. Test the function for this file and different values ​​of `n`.

````python
# Solution of exercise 3
import csv
def zad3(f, n):
    try:
        with open(f, 'r',  encoding='utf-8') as file:
            csv_reader = csv.reader(file)
            for i in range(n):
                row = next(csv_reader, None)
                if row is not None:
                    print(*row)
                else:
                    break
    except Exception as e:
        print(f"Error reading CSV file: {e}")
f = "bwq.csv"
n = int(input())
zad3(f, n)

# newline=''
````

````python
# Solution of exercise 4
import csv
def file(f, n):
    try:
        with open(f, 'r',  encoding='utf-8') as file:
            csv_reader = csv.reader(file)
            header = next(csv_reader)
            for i, row in enumerate(csv_reader, start=2):
                try:
                    date_time_str = row[0]
                    temperature = float(row[2])
                    if temperature >= n:
                        beach_name = row[0]
                        measuring = row[1]
                        print(f"{i}, {temperature}, {beach_name}, {measuring}")
                except (ValueError, IndexError) as e:
                    print(f"Error {i}: {e}")
    except Exception as e:
        print(f"Error reading CSV file: {e}")
f = "bwq.csv"
n = int(input())
file(f, n)
````

````python
# Solution of exercise 5
import pandas as pd
def zad5(file, n):
    try:
        df = pd.read_csv(file, encoding='utf-8', nrows=n)
        display(df)
        print(df.to_string(index=False, header=False))
    except Exception as e:
        print(f"Error reading the CSV file: {e}")
file = "bwq.csv"
n = int(input())
zad5(file, n)
````
