Execute the following code and pay attention to the error message.
```python
a = 1
b = 0
x = a / b
print(x)
```

Expand the code so that, regardless of the values of `a` and `b`, the program does not crash and display the error message generated by the Python interpreter. Instead, the program should detect that division is not possible and display its own message, such as:
```
Something wrong with division...
```

You might find this page useful: https://www.w3schools.com/python/python_try_except.asp

## Task 2

Execute the code
```python

```
first for the assignment `a = 1`, then `a = "1"`. Pay attention to the **exception names** that appear in the error messages. **Use them** to modify the code from the previous task so that instead of the text
```
Something wrong with division...
```
more detailed information appears, i.e. one of the following three messages should appear, depending on the situation:
```
Division by zero!!!
```
or
```
Unsupported operand types for division!!!
```
or
```

```


````python
# Solution of exercise 2.
a = 1
b = 0
try:
  x = a / b
  print(x)
except ZeroDivisionError:
  print("Division by zero!!!")
except TypeError:
  print("Unsupported operand types for division!!!")
except:
  print("Something else wrong with division...")
print()

a = "1"
b = 0
try:
  x = a / b
  print(x)
except ZeroDivisionError:
  print("Division by zero!!!")
except TypeError:
  print("Unsupported operand types for division!!!")
except:
  print("Something else wrong with division...")
````

## Task 5

Write a function `get_float` that asks the user to enter a floating-point number and returns its value (a floating-point number, not a string). If the user does not enter a floating-point number (e.g., a string), the function displays the message `"Incorrect value."` and returns the default value 1.

This is a function analogous to the function from task 6 of labs 6, but for floating-point numbers, not integers.

````python
# Solution of exercise 5.
def get_float(promt):
  liczba = input(promt)
  try:
    float(liczba)
    return liczba
  except:
      print("Incorrect value.")
      return 1
print(get_float("Enter a float number:"))

````

## Task 6

Modify the function from the previous task so that the user is asked to enter a floating-point number until he finally makes a mistake and enters it.

````python
# Solution of exercise 6.
def get_float2(promt):
  while True:
    liczba = input(promt)
    try:
      float(liczba)
      return liczba
    except:
        print("Incorrect value.")
print(get_float2("Enter a float number:"))
````
